```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
from scipy.integrate import solve_ivp

from IPython.display import display, Latex, Math
```

# Task 3


```python
def oscillator(t, y):
    y1, y2 = y
    dy1dt = y2                                      # y' = y2
    dy2dt = (-5 * y2 - 500 * y1 - 10000 * y1 ** 3)  # y''
    return [dy1dt, dy2dt]

# Initial condition
t_eval = np.linspace(0, 5, 5000)

solution = solve_ivp(
    fun=oscillator, 
    t_span=[0, 5], 
    y0=[0.02, 0.0],
    method='RK45',
    t_eval=t_eval,
    rtol=1e-6
)

t = solution.t
y = solution.y[0]
ydot = solution.y[1]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[14, 4])

for ax in [ax1, ax2]:
    ax.grid(True, which="major", linestyle="--", linewidth=.75, alpha=.75)
    
ax1.set_xlabel(r"$t$", fontsize=14)
ax1.set_ylabel(r"$x\;\!(t)$", fontsize=14)
ax1.set_title('Oscillator\'s displacement')

ax1.xaxis.set_minor_locator(ticker.MultipleLocator(0.2))
ax1.yaxis.set_minor_locator(ticker.MultipleLocator(0.05))

ax1.plot(
    t, y,
    color='royalblue'
)

ax2.set_xlabel(r"$x\;\!(t)$", fontsize=14)
ax2.set_ylabel(r"$x'\;\!(t)$", fontsize=14)
ax2.set_title('Phase portrait')

ax2.xaxis.set_minor_locator(ticker.MultipleLocator(0.001))
ax2.yaxis.set_minor_locator(ticker.MultipleLocator(0.02))

ax2.plot(
    y, ydot,
    color=plt.cm.tab10(1)
)

plt.show()
```
